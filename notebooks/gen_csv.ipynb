```python
from utils import generate_plot_catalog
import pandas as pd
```

```python
df = dict()
for casename in [
    "g.e22b05.G1850ECOIAF_JRA.TL319_g17.cocco.001",
    "g.e22.G1850ECO_JRA_HR.TL319_t13.003",
    "g.e22.G1850ECO_JRA_HR.TL319_t13.004",
]:
    df[casename] = pd.DataFrame({})
    for plottype in [
        "histogram",
        "summary_map",
        "time_series",
        "trend_hist",
        "trend_map",
    ]:
        print(f"Generating csv for {plottype} plots in {casename}")
        df[casename] = pd.concat(
            [
                df[casename],
                generate_plot_catalog(
                    "./",
                    image_dir_name=f"images/{casename}/{plottype}",
                    use_full_path=False,
                ),
            ]
        )
    df[casename].to_csv(
        f"images/{casename}/png_catalog.csv",
        compression=None,
        index=False,
    )
```
